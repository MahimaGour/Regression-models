### Father Son Hieght DataSet for LINEAR REGRESSION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('newfile.csv')
data.head()
```

```python
x = data['Father'].values.reshape(-1,1)
y = data['Son'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
plt.scatter(X_train, y_train, c="lightblue")
plt.plot(X_train, lr.predict(X_train), c='blue', linewidth=4)
plt.show()
```

```python
plt.scatter(X_test, y_test, c="lightblue")
plt.plot(X_test, lr.predict(X_test), c='blue', linewidth=4)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

print("MSE OF TRAIN DATA: ",mean_squared_error(X_train, lr.predict(X_train)))

print("MSE OF TEST DATA: ",mean_squared_error(X_test, lr.predict(X_test)))
```

```python
import math

from sklearn.metrics import mean_squared_error
print("RMSE OF TRAIN DATA: ",math.sqrt(mean_squared_error(X_train, lr.predict(X_train))))

print("RMSE OF TEST DATA: ",math.sqrt(mean_squared_error(X_test, lr.predict(X_test))))
```
